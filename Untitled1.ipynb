```python
#import #  
import time
from selenium import webdriver
# import chrome in webdriver
from selenium.webdriver import Chrome
#import pandas 
import pandas as pd
#import selenium webdriver webdriver
from selenium.webdriver.common.keys import Keys as key
import warnings
warnings.filterwarnings('ignore')
```

```python
#open webpage in chrome
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get('https://www.hostelworld.com/')
doc.maximize_window()
```

```python
# Enter location
doc.find_element_by_id('location-text-input-field').click()
time.sleep(2)
doc.find_element_by_id('search-input-field').send_keys('London')
time.sleep(3)
doc.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[2]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div').click()
time.sleep(3)
#click let's go
doc.find_element_by_css_selector('button[id="search-button"]').click()
```

```python
#blank list variable
URLS = []
hostel_name=[]
distance=[]
ratings=[]
total_reviews=[]
privates=[]
dorms=[]
facilities=[]
description=[]


#loop run 30 times
for i in range(1,31):
    try:
        URLS.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//h2[@class="title title-6"]//a').get_attribute('href'))
    except:
        URLS.append('-')
    
    try:
        hostel_name.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//h2[@class="title title-6"]//a').text)
    except:
        hostel_name.append('-')
    
    try:
        distance.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//span[@class="description"]').text)
    except:
        distance.append('-')
    
    try:
        ratings.append(doc.find_element_by_xpath(f'//div[@class="property-card"][2]//div[@class="info"]//a').text.split('\n')[0])#rating
    except:
        ratings.append('-')
     
    try:
        total_reviews.append(doc.find_element_by_xpath(f'//div[@class="property-card"][2]//div[@class="info"]//a').text.split('\n')[2])#Total review)
    except:
        total_reviews.append('-')
    
    try:
        privates.append(doc.find_element_by_xpath(f'//div[@class="property-card"][1]//div[@class="price-col"]').text.split('\n')[-1])#privates form
    except:
        privates.append('-')
                        
    try:
        dorms.append(doc.find_element_by_xpath(f'//div[@class="property-card"][2]//div[@class="price-col"][2]').text.split('\n')[-1])#Dorms form)
    except:
        dorms.append('-')            
print('scrape successfully')

#scrape description and facilities
for i in URLS:
    doc.get(i)
    time.sleep(2)
    try:
        facilities.append(doc.find_element_by_xpath('//div[@class="facilities-container"]//ul[@class="facilities"]').text)
    except:
        facilities.append('-')
    
    try:
        description.append(doc.find_element_by_xpath('//div[@class="content collapse-content"]').text.replace('\n',' '))
    except:
        description.append('-')
        
print('scrape successfully')
```
