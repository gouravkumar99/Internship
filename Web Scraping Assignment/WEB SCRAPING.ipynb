# 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp .

```python
# defind a variable for link
url = r'https://meesho.com/bagsladies/pl/p7vbp'
 
# defind a variable for get all data to variable
# using requests all html data to r
r = requests.get(url)


soup = bs(r.content, 'html.parser')
```

```python
# product name in p tags
soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0')# find_all p tags
```

```python
print(soup.find_all('p')[0].text)#using slicing to slice product name price and discount then extract text in slicing
print(soup.find_all('p')[1].text)#price
print(soup.find_all('p')[2].text)#Discount
```

```python
# create list variable
product_list = []#for product list 
discount_list=[]#for discount list 
count = 0
# count class of p tag and for_loop run 
for i in range(0,len(soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0'))):
    if count<len(soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0')):#use if condition to check count variable no greater then tags
        product_list.append(soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0')[count].text)#scrape product list using slicing
        discount_list.append(soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0')[count-1].text)#scrape discount list using slicing
    count +=3 # add 3 because procud name gap 3 times 
product_list = product_list[:-1]#using slicing for remove last text
discount_list = discount_list[1:]#using slicing for remove 1st text
```

```python
#create list variable and use h5 heading  tag beuause all price data in h5 heading  tags
price_data = soup.find_all('h5')
```

```python
price_list = []#create list variable
#for_loop run price data len 
for i in range(len(price_data)):
    price_list.append(price_data[i].text)#all data add in list using for_loop 
price_list   
```

```python
print(f'Product list- \n{product_list}\n\n Product price list- \n{price_list}\n\n Product discount list-\n{discount_list}')
```

```python
print('''############################################
#     product name, price and discounts    #
############################################''')
pd.DataFrame({"Product list":product_list,"Product price list":price_list,"Product discount list":discount_list})
```

```python
#create a variable and use class to tag all website data in this class of tag  
data = soup.find_all('div',class_='site-container')
```

```python
#create a list varable 
all_heading = []
#use for_loop and run 6 times beacause 6 having tags in html
for i in range(1,7):
    all_heading.append(data[0].find_all(f'h{i}'))#using loop and heading tags all heading data append into list
print(all_heading)#Pring all hading data
```

```python
headings = []#create blank list variable 
headings.append(all_heading[0][0].text)#using slicing extract text then append h1 heading data into list
print(headings)
```

```python
# for_loop run using slicing len
for i in range(len(all_heading[1])):
    headings.append(all_heading[1][i].text)#using slicing and extract h2 heading text then append into list
    
print(headings)
```

```python
# for_loop run using slicing len
for i in range(len(all_heading[3])):
    headings.append(all_heading[3][i].text)#using slicing and extract h2 heading text then append into list
```

```python
print('''########################
#   All heading   #
########################''')
print(f'\n{headings}')
```

```python
# create variable 
all_date = soup.find_all('p',class_='entry-meta')#use class of p tag because all date data in this tags
```

```python
date = []#create blank list variable

#for_loop run data len time
for i in range(0,len(all_date),2):
    #using slicing extract text and replace unnecessary text to blank and using stip remove unnecessary space 
    date.append(all_date[i].text.replace('by Corey Schafer Leave a Comment','').strip())

    
```

```python
print('''########################
#       All date       #
########################''')
print(f'\n{date}')
```

```python
all_content = soup.find_all('div',class_='entry-content')
# all_content[0].text
```

```python
content = []#create blank list variable
# run for_loop all_content len times
for i in range(len(all_content)):
    #using slicing and extract text then split data new line and then append into list
    content.append(all_content[i].text.split('\n')[1])
```

```python
print('''########################
#      all content     #
########################''')
#loop run content len times    
for i in range(len(content)): 
    #final all content print new line 
    print(f'\n{content[i]}')
```

```python
# create variable 
all_link = soup.find_all('span',class_='embed-youtube')#use class of span tag all link in this tag 
print(all_link)
```

```python
# create blank list variable 
YouTub_link = []
# run for_loop all_link len times
for i in range(len(all_link)):
    YouTub_link.append(all_link[i].iframe['src'])#using slicing extract data then extract link using iframe 
print(YouTub_link)   
```

```python
print('''########################
#    All YouTub link   #
########################''')
#loop run YouTub_link len times    
for i in range(len(YouTub_link)): 
    #all YouTub_link print new line 
    print(f'\n{YouTub_link[i]}')
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
    i) Restaurant name
    ii) Cuisine
    iii) Location
    iv) Ratings
    v) Image URL 

```python
# defind a variable for link
url = r'https://www.dineout.co.in/delhi-restaurants/buffet-special'
 
# defind a variable for get all data to variable
# using requests all html data to r

r = requests.get(url)
soup = bs(r.content, 'html.parser')
```

```python
restarrant_data = soup.find_all('div',class_='restaurant-info-section restnt-detail')
img_data = soup.find_all('div',class_='img-wrap')
```

```python
Restaurant_name = []
Cuisine = []
Location = []
Ratings = []
Image_URL  = []

for i in range(len(restarrant_data)):
    Restaurant_name.append(restarrant_data[i].find_all('a')[0].text)  #name
    Cuisine.append(restarrant_data[i].find_all('div',class_='restnt-loc ellipsis')[0].text) #location
    Location.append(restarrant_data[i].find_all('span',class_='double-line-ellipsis')[0].text.split('|')[1]) #Cuisine
    Ratings.append(img_data[i].find_all('div',class_='restnt-rating rating-4')[0].text) #rating
    Image_URL.append(img_data[i].find_all('img',class_='no-img')[0]['data-src']) #image link
```

```python
print('''#####################
#  Restaurant name  #
#####################\n''')
#loop run Restaurant_name len times    
for i in range(len(Restaurant_name)): 
    #all Restaurant_name print new line 
    print(f'{Restaurant_name[i]}')
```

```python
#####################
#       Cuisine     #
#####################
```

```python
print('''#####################
#       Cuisine     #
#####################\n''')
#loop run Restaurant_name len times    
for i in range(len(Cuisine)): 
    #all Cuisine print new line 
    print(f'{Cuisine[i]}')
```

```python
print('''#####################
#       Location    #
#####################\n''')
#loop run Location len times    
for i in range(len(Location)): 
    #all Location print new line 
    print(f'{Location[i]}')
```

```python
print('''#####################
#       Ratings     #
#####################\n''')
#loop run Ratings len times    
for i in range(len(Ratings)): 
    #all Ratings print new line 
    print(f'{Ratings[i]}')
```

```python
print('''#####################
#      Image URL    #
#####################\n''')
#loop run Image URL len times    
for i in range(len(Image_URL)): 
    #all Image URL print new line 
    print(f'{Image_URL[i]}')
```
