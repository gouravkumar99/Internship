# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
    Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
    You need to find following details:
    A) Rank
    B) Name
    C) Artist
    D) Upload date
    E) Views

```python
#open web
url = r'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#reate blank list variable
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []

#loop run 30 time
for i in range(1,31):
    #extract data and data append 
    Rank.append(doc.find_element_by_xpath(f'//table[@class="wikitable sortable jquery-tablesorter"]//tr[{i}]//td[1]').text)
    Name.append(doc.find_element_by_xpath(f'//table[@class="wikitable sortable jquery-tablesorter"]//tr[{i}]//td[2]').text)
    Artist.append(doc.find_element_by_xpath(f'//table[@class="wikitable sortable jquery-tablesorter"]//tr[{i}]//td[3]').text)
    Upload_date.append(doc.find_element_by_xpath(f'//table[@class="wikitable sortable jquery-tablesorter"]//tr[{i}]//td[5]').text)
    Views.append(doc.find_element_by_xpath(f'//table[@class="wikitable sortable jquery-tablesorter"]//tr[{i}]//td[4]').text)
```

```python
#creat variable and make dataframe 
df = pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'Upload date':Upload_date,'Views':Views})
#check first 5 rows 
df.head()
```

```python
#open web
url = r'http://statisticstimes.com/'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#create blank list variable
Rank =[]
State=[]
GSDP1=[]
GSDP2=[]
Share=[]
GDP=[]
#create variable 
i = 1

#run while loop if data not match so loop break
while True:
    try:
        Rank.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[1]').text)
        State.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[2]').text)
        GSDP1.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[3]').text+" | "+doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[4]').text)
        GSDP2.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[7]').text+" | "+doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[8]').text)  
        Share.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[5]').text)
        GDP.append(doc.find_element_by_xpath(f'//table[@id="table_id"]//tbody//tr[{i}]//td[6]').text)
        i+=1
    except:
        break
print("scrape successfully")
```

```python
#create data frmame
data = pd.DataFrame({'Rank':Rank,'State':State,'GSDP (Cr INR at Current prices)':GSDP1,'GSDP (Cr INR at 2011-12 prices)':GSDP2,'Share':Share,'GDP':GDP})
#check first five rows of data framke
data.head()
```

# 5. Scrape the details of trending repositories on Github.com.
    Url = https://github.com/
    You have to find the following details:
    A) Repository title
    B) Repository description
    C) Contributors count
    D) Language used
    Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#open web
url = r'https://github.com/'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#open trending
doc.get(doc.find_element_by_xpath('//ul[@class="d-lg-flex list-style-none"]//li[4]//ul[@class="list-style-none f5 pb-1"]//li[5]//a').get_attribute('href'))
```

```python
# create blank list variable
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Contributors_url=[]
Language_used=[]

#create variable
i=1
# loop run condition true tiem
while True:
    try:
        Repository_title.append(doc.find_element_by_xpath(f'//article[@class="Box-row"][{i}]//h1').text)
        try:
            Repository_description.append(doc.find_element_by_xpath(f'//article[@class="Box-row"][{i}]//p').text)
        except:
            Repository_description.append('-')
        Contributors_url.append(doc.find_element_by_xpath(f'//div//article[@class="Box-row"][{i}]//h1[@class="h3 lh-condensed"]//a').get_attribute('href'))
        try:
            Language_used.append(doc.find_element_by_xpath(f'//div//article[@class="Box-row"][{i}]//div[@class="f6 color-fg-muted mt-2"]//span[@class="d-inline-block ml-0 mr-3"]').text)
        except:
            Language_used.append('-')
        i+=1
    except:
        #loop exit
        break
print("scrape successfully")

#loop run Contributors_url len
for i in Contributors_url:
    #open page url
    doc.get(i)
    time.sleep(2)
    try:
        #append data into list
        Contributors_count.append(doc.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]').text.split('Contributors')[1].split('\n')[0])     
    except:
        
        Contributors_count.append('-')
print("scrape successfully")        
        
```

```python
pd.DataFrame({'Repository title':Repository_title,'Repository_description':Repository_description,'Language_used':Language_used,'Contributors count':Contributors_count})
```

```python
#open web
url = r'https://archive.ics.uci.edu/'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#click view all dataset
doc.find_element_by_xpath('//span[@class="whitetext"][2]//a[1]').click()
time.sleep(2)

##blank list variable
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

#create variable
i = 1

#loop run condition true time
while True:
    try:
        #append data into list
        Dataset_name.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[1]').text)
        Data_type.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[2]').text)
        Task.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[3]').text)
        Attribute_type.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[4]').text)
        No_of_instances.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[5]').text)
        No_of_attribute.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[6]').text)
        Year.append(doc.find_element_by_css_selector('table[cellpadding="3"]').find_element_by_xpath(f'//td[2]//tr[{i}]//td[7]').text)
        #add 1 value
        i+=1
    except:
        #break code
        break

print("scrape successfully")
        
```

```python
#create DataFrame
pd.DataFrame({'Dataset name':Dataset_name,'Data type':Data_type,'Task':Task,'Attribute type':Attribute_type,'No of instances':No_of_instances,'No of attribute':No_of_attribute,'Year':Year})    
```
