# 1) Write a python program to display all the header tags from wikipedia.org

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# create function 
def html_data(url):
    # using requests all html data to r
    r = requests.get(url)
    all_html = bs(r.content, 'html.parser')
    return all_html

#create variable and all html data in this variable 
all_html = html_data(r'https://en.wikipedia.org/wiki/Main_Page')
```

```python
#for_loop run 6 times because 6 headig tag in html
for i in range(1,7):
    print(all_html.find_all(f'h{i}'))#print all heading tags 
```

```python
#create balnk list variable
all_heading = []
#for_loop run h1 tag len time
for i in range(len(all_html.find_all('h1'))):
#     using slicing extract text the append into list
    all_heading.append(all_html.find_all('h1')[i].text)

#for_loop run h2 tag len time
for i in range(len(all_html.find_all('h2'))):
#     using slicing extract text the append into list
    all_heading.append(all_html.find_all('h2')[i].text)

#for_loop run h2 tag len time
for i in range(len(all_html.find_all('h3'))):
#     using slicing extract text then split data then append into list
    all_heading.append(all_html.find_all('h3')[i].text.split('\n')[1])
```

```python
#create data frame using pandas
pd.DataFrame({'All header tags':all_heading})
```

# # 2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
# defind a variable for link
url = r'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
#create variable and all html data in this variable 
moviedata_page1 = html_data(url)
```

```python
# create 3 list variable 
movies_list = []
year_list = []
rating_list = []

# create a function because all pages same html
def movie_data(movie_page):
    #use strong tag all rating in this tag then extract data then append into list
    for n in range(len(movie_page.find_all('strong')[2:])):
        rating_list.append(movie_page.find_all('strong')[2:][n].text)
        
    # using for_loop and loop run top_100_movies of len times
    for i in range(len(movie_page.find_all('h3'))-1):#last value is blank in list so use -1
        movies_list.append(movie_page.find_all('h3')[i].text.split('\n')[2])
        year_list.append(movie_page.find_all('h3')[i].text.split('\n')[3])
        
        
movie_data(moviedata_page1)
```

```python
# 1st page only 50 movies so use 2nd page url 
url = r'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt'

# using requests all html data to r
r = requests.get(url)
#create variable all html data in this variable 
moviedata_page2 = bs(r.content, 'html.parser')

#funcion run and append 2 paga movie data 
movie_data(moviedata_page2)
```

```python
# print all data in data frame
print('''###################################################################################
#      IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)     #
###################################################################################''')

#make data frame using pandas all data in data frame
pd.DataFrame({'Movies Name':movies_list,'Movies Rating':rating_list,'Movies year of release':year_list})
```

```python
#for_loop run price data len 
for i in range(len(product_data.find_all('h5'))):
    price_list.append(product_data.find_all('h5')[i].text)#all data add in list using for_loop 
len(price_list )  
```

```python
print('''#########################
#  All Product Details  #
#########################''')

# Use pandas for all list data to Dataframe 
# use pandas DataFrame function
pd.DataFrame({'product name':product_list,'Discount':discount_list,'Price':price_list,})
```
