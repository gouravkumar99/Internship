```python
URL = []
for r in range(3):
    urls=doc.find_elements_by_xpath("//a[@class='a-link-normal s-link-style a-text-normal']")#collecting urls of all the laptop
    for i in urls:
        URL.append(i.get_attribute('href'))
    
    #click next page
    doc.find_element_by_xpath('//span[@class="s-pagination-strip"]//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
    time.sleep(3)
    
# blank list variables   
product = []
price = []
brand = []
exchange = []
availability = []
delivery = []

for url in URL:##loop for every guitar in the list
    doc.get(url)
    time.sleep(2)
    try:
        product.append(doc.find_element_by_xpath('//span[@id="productTitle"]').text)
    except:
        product.append('-')
    try:
        try:
            price.append(doc.find_element_by_xpath('//td[@class="a-span12"]').text)
        except:
            price.append(doc.find_element_by_css_selector('span[class="a-price-whole"]').text)             
    except:
        price.append('-')
    try:
        brand.append(doc.find_element_by_xpath('//div[@class="a-section a-spacing-small a-spacing-top-small"]//span[@class="a-size-base"]').text)
    except:
        brand.append('-')
    try:
        exchange.append(doc.find_element_by_id('RETURNS_POLICY').text)
    except:
        exchange.append('-')
    try:
        availability.append(doc.find_element_by_id('availability').text)
    except:
        availability.append('-')
    try:
        delivery.append(doc.find_element_by_id('ddmDeliveryMessage').text)
    except:
        delivery.append('-')
    
print('successful scrap')
```

# 3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
#open web
url = r'https://images.google.com/'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#create image scrapinf function
def scrape_image(search):
    try:
        doc.find_element_by_css_selector('input[class="gLFyf gsfi"]').clear()
    except:
        doc.find_element_by_class_name('og3lId').clear()
    time.sleep(2)
    try:
        doc.find_element_by_css_selector('input[class="gLFyf gsfi"]').send_keys(search)
    except:
        doc.find_element_by_class_name('og3lId').send_keys(search)
    try:
        doc.find_element_by_class_name('zgAlFc').click()
    except:
        doc.find_element_by_class_name('XZ5MVe').click()
    time.sleep(2)
```

```python
#create blank list variable
fruits=[]
#use function scrape 10 images
scrape_image('fruits')
for i in range(1,11):
    fruits.append(doc.find_element_by_xpath(f'//*[@id="islrg"]/div[1]/div[{i}]//img').get_attribute('src'))
```

```python
cars=[]
#use function scrape 10 images
scrape_image('cars')
for i in range(1,11):
    cars.append(doc.find_element_by_xpath(f'//*[@id="islrg"]/div[1]/div[{i}]//img').get_attribute('src'))
```

```python
Machine_Learning=[]
#use function scrape 10 images
scrape_image('Machine Learning')
for i in range(1,11):
    Machine_Learning.append(doc.find_element_by_xpath(f'//*[@id="islrg"]/div[1]/div[{i}]//img').get_attribute('src'))
```

```python
Guitar=[]
#use function scrape 10 images
scrape_image('Guitar')
for i in range(1,11):
    Guitar.append(doc.find_element_by_xpath(f'//*[@id="islrg"]/div[1]/div[{i}]//img').get_attribute('src'))
```

```python
Cakes=[]
#use function scrape 10 images
scrape_image('Cakes')
for i in range(1,11):
    Cakes.append(doc.find_element_by_xpath(f'//*[@id="islrg"]/div[1]/div[{i}]//img').get_attribute('src'))
```

```python
pd.DataFrame({'fruits':fruits,'cars':cars,'Machine Learning':Machine_Learning,'Guitar':Guitar,'Cakes':Cakes})
```

# 7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
#open web
url = r'https://www.digit.in/search/?keyword=best%20gaming%20laptops'
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get(url)
doc.maximize_window()
```

```python
#create blank list variable
URLS = []
for i in range(10):
    URLS.append(doc.find_element_by_xpath(f'//div[@class="searchPage"][{i+1}]//a').get_attribute('href'))
```

```python
#create blank list variable
Product_name=[]
Market_Status=[]
Release_Date=[]
Official_Website=[]
price=[]

#loop run in urls len
for i in URLS:
    doc.get(i)#open url
    time.sleep(3)#wait 3 sec
    try:
        Product_name.append(doc.find_element_by_xpath('//div[@class="heading-wraper"]//h1').text)
    except:
        Product_name.append('-')
        
    try:
        Market_Status.append(doc.find_element_by_xpath('//div[@id="overview"]//div[@class="status_bar"]').text.split(':')[1].split('\n')[0])
    except:
        Market_Status.append('-')
        
    try:
        Release_Date.append(doc.find_element_by_xpath('//div[@id="overview"]//div[@class="status_bar"]').text.split(':')[2].split('\n')[0])
    except:
        Release_Date.append('-')
        
    try:
        Official_Website.append(doc.find_element_by_xpath('//div[@id="overview"]//div[@class="status_bar"]').text.split(':')[3].split('\n')[0])
    except:
        Official_Website.append('-')
        
    try:
        price.append(doc.find_element_by_xpath('//div[@id="overview"]//div[@class="price"]//strong').text)
    except:
        price.append('-')
    
print('successful scrap')
```

```python
pd.DataFrame({'Product_name':Product_name,'Market_Status':Market_Status,'Release_Date':Release_Date,'Official_Website':Official_Website,'price':price,'URL':URLS})
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
#open webpage in chrome
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get('https://www.youtube.com/watch?v=3gk3_EkVUUQ')
doc.maximize_window()
```

```python
#create blank list variable
name = []
Comment = []
upvote = []
time_ = []

#use for page scroling
page = doc.find_element_by_tag_name('html')
for i in range(1,520):
    try:
        name.append(doc.find_element_by_xpath(f'//div[@id="contents"]//ytd-comment-thread-renderer[@class="style-scope ytd-item-section-renderer"][{i}]//span[@class="style-scope ytd-comment-renderer"]').text)
    except: 
        page.send_keys(key.END)#if error page scrole down
        while True:
            try:
                name.append(doc.find_element_by_xpath(f'//div[@id="contents"]//ytd-comment-thread-renderer[@class="style-scope ytd-item-section-renderer"][{i}]//span[@class="style-scope ytd-comment-renderer"]').text)
                break
            except:
                continue
    try:
        Comment.append(doc.find_element_by_xpath(f'//div[@id="contents"]//ytd-comment-thread-renderer[@class="style-scope ytd-item-section-renderer"][{i}]//yt-formatted-string[@id="content-text"]').text)
    except:
        Comment.append('-')
    try:
        upvote.append(doc.find_element_by_xpath(f'//div[@id="contents"]//ytd-comment-thread-renderer[@class="style-scope ytd-item-section-renderer"][{i}]//span[@id="vote-count-middle"]').text)
    except:
        upvote.append('-')
    try:
        time_.append(doc.find_element_by_xpath(f'//div[@id="contents"]//ytd-comment-thread-renderer[@class="style-scope ytd-item-section-renderer"][{i}]//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]').text)
    except:
        time_.append('-')
        
print('successful scrap')
```

```python
pd.DataFrame({'name':name,'Comment':Comment,'upvote':upvote,'time_':time_})
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
#open webpage in chrome
doc = webdriver.Chrome(executable_path='chromedriver.exe')
doc.get('https://www.hostelworld.com/')
doc.maximize_window()
```

```python
# Enter location
doc.find_element_by_id('location-text-input-field').click()
time.sleep(2)
doc.find_element_by_id('search-input-field').send_keys('London')
time.sleep(3)
doc.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[2]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div').click()
time.sleep(3)
#click let's go
doc.find_element_by_css_selector('button[id="search-button"]').click()
```

```python
#blank list variable
URLS = []
hostel_name=[]
distance=[]
ratings=[]
total_reviews=[]
privates=[]
dorms=[]
facilities=[]
description=[]


#loop run 30 times
for i in range(1,31):
    try:
        URLS.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//h2[@class="title title-6"]//a').get_attribute('href'))
    except:
        URLS.append('-')
    
    try:
        hostel_name.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//h2[@class="title title-6"]//a').text)
    except:
        hostel_name.append('-')
    
    try:
        distance.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//span[@class="description"]').text)
    except:
        distance.append('-')
    
    try:
        ratings.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//div[@class="info"]//a').text.split('\n')[0])#rating
    except:
        ratings.append('-')
     
    try:
        total_reviews.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//div[@class="info"]//a').text.split('\n')[2])#Total review)
    except:
        total_reviews.append('-')
    
    try:
        privates.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//div[@class="price-col"]').text.split('Rs')[-1])#privates form
    except:
        privates.append('-')
                        
    try:
        dorms.append(doc.find_element_by_xpath(f'//div[@class="property-card"][{i}]//div[@class="price-col"][2]').text.split('Rs')[-1])#Dorms form)
    except:
        dorms.append('-')            
print('scrape successfully')

#scrape description and facilities
for i in URLS:
    doc.get(i)
    time.sleep(2)
    try:
        facilities.append(doc.find_element_by_xpath('//div[@class="facilities-container"]//ul[@class="facilities"]').text)
    except:
        facilities.append('-')
    
    try:
        description.append(doc.find_element_by_xpath('//div[@class="content collapse-content"]').text.replace('\n',' '))
    except:
        description.append('-')
        
print('scrape successfully')
```

```python
print('Data for all available Hostels')
pd.DataFrame({'hostel_name':hostel_name,'distance from city centre':distance,'Ratings':ratings,'Total reviews':total_reviews,'privates from price':privates,' dorms from price':dorms,'facilities':facilities,'description':description})
```
